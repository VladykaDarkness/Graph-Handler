```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

path = 'waveforms/060924/Осциллограммы переходов/shot13.csv'
# path = 'waveforms/110924/Tek011_ALL.csv'
df = pd.read_csv(path)
print(df)
print(path)
inf_indexes = df[np.isinf(df['CH3V'])].index.tolist()
print(inf_indexes)
df.replace([np.inf, -np.inf], np.nan, inplace=True)
inf_indexes = df[np.isinf(df['CH3V'])].index.tolist()
print(inf_indexes)
print(df['CH3V'].max()*1e3)
# files = sorted(os.listdir("waveforms/11"))
# print(files)
# csv_files = []

# for file in files:
#     if (file.endswith('.csv')):
#         csv_files.append(file)
# print(csv_files)
def plot_graph():
    plt.figure(figsize=(10, 6))
    plt.rcParams["font.family"] = "serif" # font
    plt.subplot(2, 1, 1)
    plt.xlim(df.iloc[0, 0]*1e9, df.iloc[-1, 0]*1e9)
    plt.ylabel('Intensity, mV')
    plt.xlabel('Time, ns')
    plt.plot(df['Time(s)']*1e9, df['CH1V']*1e3, color='red')

    plt.subplot(2, 1, 2)
    plt.xlim(df.iloc[0, 0]*1e9, df.iloc[-1, 0]*1e9)
    # plt.ylim(0, df['CH2V'].mean()*1e3 * 2 )
    plt.ylabel('Reflectance, of the GST225, mV')
    plt.xlabel('Time, ns')
    plt.plot(df['Time(s)']*1e9, df['CH3V']*1e3)

    plt.tight_layout()  # Автоматическая регулировка расстояния между графиками

plot_graph()
plt.show()
# save = input('Save?(y or n): ')
# if(save == 'y'): 
#     plt.savefig('006.png', dpi=300, bbox_inches='tight')
```
